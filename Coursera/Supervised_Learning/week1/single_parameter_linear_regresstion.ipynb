```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')
from lab_utils_uni import plt_house_x, plt_contour_wgrad, plt_divergence, plt_gradients
```

```python
# create simple dataset
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
# get the number of data
m=x_train.shape[0]
```

```python
# function to compute lost
def compute_lost(x, y, w, b):
    lost = 0
    for i in range(m):
        f_wb = w*x_train[i] + b
        lost = lost + (f_wb - y[i])**2
    avg_lost = 1/(2*m)*lost
    return avg_lost
```

```python
# function to compute gradient
def compute_gradient(x, y, w, b):
    # calculate derivative of lost function
    dj_dw = 0
    dj_db = 0
    for i in range(m):
        dj_dw_i = 2*(w*x[i] + b - y[i])*x[i]
        dj_dw += dj_dw_i
        dj_db_i = 2*(w*x[i] + b )- y[i]
        dj_db += dj_db_i
    return dj_dw/m, dj_db/m
```

```python
plt_gradients(x_train,y_train, compute_lost, compute_gradient)
plt.show()
```
